# Data Visualization Analysis Types

First, let's prepare a sample dataset using `seaborn`.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the 'tips' dataset
df = sns.load_dataset('tips')
display(df.head())
```

### Filtered Correlation Analysis: Dinner Time

Let's see if the relationships between variables change when we only look at dinner data.

```python
# Filter for Dinner
dinner_df = df[df['time'] == 'Dinner']

# Recalculate correlation matrix for numeric columns
dinner_corr = dinner_df.select_dtypes(include=['float64', 'int64']).corr()

print("Correlation Matrix for Dinner Time:")
display(dinner_corr)

# Visualize with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(dinner_corr, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap (Dinner Only)')
plt.show()
```

### 1. Univariate Analysis
This looks at one feature at a time. We use this to understand the distribution, central tendency, and outliers of a specific column.

```python
plt.figure(figsize=(10, 5))

# Histogram with KDE for Univariate Analysis
sns.histplot(df['total_bill'], kde=True, color='skyblue')
plt.title('Univariate Analysis: Distribution of Total Bill')
plt.show()
```

### 2. Bivariate Analysis
This examines the relationship between two variables. For example, how does the `total_bill` affect the `tip`?

```python
plt.figure(figsize=(10, 5))

# Scatter plot for Bivariate Analysis
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.title('Bivariate Analysis: Total Bill vs Tip')
plt.show()
```

### 3. Multivariate Analysis
This involves three or more variables. We can use color (`hue`), size, or different facets to introduce more dimensions.

```python
plt.figure(figsize=(10, 6))

# Scatter plot with hue and style for Multivariate Analysis
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.title('Multivariate Analysis: Bill vs Tip by Day, Time, and Table Size')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

### 5. Pair Plot (Comprehensive Analysis)

A pair plot provides a matrix of plots, showing both distributions and pairwise relationships across the entire dataset.

```python
# Create a pair plot colored by 'day'
sns.pairplot(df, hue='day', palette='viridis', corner=True)
plt.suptitle('Pairwise Relationships in Tips Dataset', y=1.02)
plt.show()
```

### 4. Correlation Heatmap (Multivariate Analysis)

Heatmaps are excellent for visualizing the correlation matrix between multiple numerical variables at once.

```python
plt.figure(figsize=(8, 6))

# Calculate correlation matrix for numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()

# Generate heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```
